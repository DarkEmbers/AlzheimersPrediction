# Feature selection

```python
features = df.drop(["Group"], axis=1)
target = df["Group"]

bestfeatures = SelectKBest(score_func=chi2, k=5)
fit = bestfeatures.fit(features, target)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(features.columns)
# concat two dataframes for better visualization
featureScores = pd.concat([dfcolumns, dfscores], axis=1)
featureScores.columns = ['Specs', 'Score']  # naming the dataframe columns
print(featureScores.nlargest(5, 'Score'))  # print n best features

print("------------")

bestfeatures = SelectKBest(score_func=f_classif, k=5)
fit = bestfeatures.fit(features, target)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(features.columns)
# concat two dataframes for better visualization
featureScores = pd.concat([dfcolumns, dfscores], axis=1)
featureScores.columns = ['Specs', 'Score']  # naming the dataframe columns
print(featureScores.nlargest(5, 'Score'))  # print n best features

print("------------")

model = RandomForestClassifier()
rfe = RFE(estimator=model, n_features_to_select=4)
rfe.fit(features, target)
support = rfe.support_
ranking = rfe.ranking_
selected_features = features.columns[support]

# Display selected features and their rankings
print("Selected Features: ", selected_features)
# If you want to see all feature rankings:
feature_ranking = pd.DataFrame({
    'Feature': features.columns,
    'Ranking': ranking
}).sort_values(by='Ranking')

print("Feature Rankings:\n", feature_ranking)
```

```python
# Import the necessary functions from sklearn.metrics
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score, roc_curve
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler


# Define target variable
target_variable = df['Group']

# Uncomment 1 of the 5 features you want to train with 
#features = df.drop(columns=['Group'])
#features = df.drop(columns=['Group','CDR'])
features = df[["MMSE", "M/F", "EDUC", "SES"]] #chi test 
#features = df[["MMSE", "M/F", "nWBV", "EDUC"]] #f-class
#features = df[["MMSE", "nWBV", "ASF", "EDUC"]] #random-forest 


scaler = StandardScaler()

scaled_features = scaler.fit_transform(features)
scaled_features_df = pd.DataFrame(scaled_features, columns=features.columns)



print("Scaled Full Features Shape:", scaled_features_df.shape)
print("Target variable name:", target_variable.name)

#scaled_features_df
#scaled_features_chi_df
#scaled_features_fclass_df
#scaled_features_rf_df

```

```python
# Split the data into training and test sets (70-30 split)
features_train_70, features_test_70, target_train_70, target_test_70 = train_test_split(
    scaled_features, target_variable, test_size=0.3, random_state=42
)

# Print the shapes of the resulting datasets to verify
print("70-30 Split - Training set shape:", features_train_70.shape, target_train_70.shape)
print("70-30 Split - Test set shape:", features_test_70.shape, target_test_70.shape)

```

```python
# Instantiate the Gaussian Naive Bayes model
model = GaussianNB()

# Train the model on the 70-30 split
model.fit(features_train_70, target_train_70)
print("Model training complete for 70-30 split.")
```

```python
#make predictions on training set 
model_train_predictions = model.predict(features_train_70)
model_train_accuracy = accuracy_score(target_train_70, model_train_predictions)

# Make predictions on the test set
model_predictictions = model.predict(features_test_70)

# Calculate accuracy and generate classification report
model_test_accuracy = accuracy_score(target_test_70, model_predictictions)
model_class_report = classification_report(target_test_70, model_predictictions)

# Display the results
print("\nPerformance for 70-30 Split:")

print("Classification Report:")
print(model_class_report)
```

**classification report analysis**
- 0 Represents Nondemented
- 1 Represents Demented
- Precision = Postive predictions made 
- Recall/sensetivity = Actual positive samples were correctly predicted
- F1-score = Combines precision and recall
- Support = Occurrences of each class (0/1) in the test set (57 Nondemented and 39 Demented)

```python
# Calculate the null accuracy for the test set
most_frequent_class = target_test_70.mode()[0]  # Find the most frequent class
null_accuracy = (target_test_70 == most_frequent_class).mean()

# Calculate the null accuracy for the test set
most_frequent_class = target_test_70.mode()[0]  # Find the most frequent class
null_accuracy = (target_test_70 == most_frequent_class).mean()

# Compare to see overfitting and underfitting
print(f"Training Accuracy (70-30 Split): {model_train_accuracy:.4f}")
print(f"Test Accuracy (70-30 Split): {model_test_accuracy:.4f}\n")

# Compare with null accuracy
print(f"Null Accuracy (Most Frequent Class): {null_accuracy:.4f}")
print(f"Model Test Accuracy: {model_test_accuracy:.4f}\n")



```

**Interpretation of Null Accuracy**
- If Model Accuracy > Null Accuracy: This indicates that the model is performing better than simply predicting the most common class, which suggests the model has learned meaningful patterns in the data.
- If Model Accuracy = Null Accuracy: This would imply that the model is not learning effectively and is close to random guessing or always predicting the most frequent class. Further improvement in model complexity or feature engineering might be needed.

**Overfitting and Underfitting**
- If training and test accuracies are close and both are high, the model is likely well-balanced and generalizes well.
- If thereâ€™s a big gap between training and test accuracy (e.g., training accuracy is much higher), it suggests overfitting.

```python
# Calculate ROC AUC
# Obtain the predicted probabilities for the positive class (Demented, coded as 1)
target_prob_70 = model.predict_proba(features_test_70)[:, 1]

roc_auc = roc_auc_score(target_test_70, target_prob_70)
print(f"ROC AUC: {roc_auc:.4f}")

# Plot the ROC Curve
fpr, tpr, thresholds = roc_curve(target_test_70, target_prob_70)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {roc_auc:.4f})")
plt.plot([0, 1], [0, 1], 'k--')  # Diagonal line for random guessing
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - 70-30 Split")
plt.legend(loc="lower right")
plt.show()
```
